# VVUQ Gate Analysis

Reads the 32-iteration index and summarizes how many candidates cleared all 10 gates and the gated composite distribution.

```python
import json
from pathlib import Path

rows = [json.loads(line) for line in Path('../data/iterations/index.jsonl').read_text().splitlines()]
cleared = sum(1 for r in rows if r['gates_all_accepted'])
scores = [r['composite_score'] for r in rows if r['composite_score'] is not None]
print(f'{cleared}/{len(rows)} iterations cleared all 10 gates')
print('composite mean', round(sum(scores)/len(scores), 3))
```
